```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
data=pd.read_csv("/content/Advertising.csv")
df=pd.DataFrame(data)
df
```

```python
#To find the summary statistics in Python we need to use the . describe() method under Pandas before that drop the column Unnamed: 0 
df1=df.drop(["Unnamed: 0"],axis=1)
df1.describe()
```

The manager wants to find out whether the same amount was spent 
for the three advertisements (TV , Radio and Newspaper).

```python
#H0=Same amount was spent for the three advertisements.
#Ha= Same amount was not spent for the three advertisements. 
#alpha=0.05
f_val,p_val=stats.f_oneway(df1["TV"],df1["Radio"],df1["Newspaper"])
print("f-value =",f_val,"p-value =",p_val)
```

The smaller the p-value, the stronger the evidence that you should reject the null hypothesis. So we can reject the null hypothesis. Therefore same amount was not spent for the three advertisements.
