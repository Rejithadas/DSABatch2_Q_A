```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("/Dataset_w6.xlsx")
print(data)
```

```python
#Find out how many countries are there in the dataset
print('The number of Countries in the dataset is',data.country.count())
```

```python
#Find out how many countries with respect to the region are available in the dataset
d1=data.groupby('Region')['country'].count()
print(d1)
```

```python
#Examine the data by using visualizations
plt.figure(figsize=(15,10))
sns.barplot(x = 'Region',y = 'GDP per capita (current US$)',data = data, palette="Blues_d")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x = 'Region',y ='International trade: Imports (million US$)',data = data, palette="Blues_d")
plt.xticks(rotation=45)
plt.show()
```

```python
#on average which region has the highest a).GDP per capita b).international trade:imports
d2=data.groupby('Region')['GDP per capita (current US$)'].mean()
print(d2)
d2.nlargest(1)
```

```python
d2=data.groupby('Region')['International trade: Imports (million US$)'].mean()
print(d2)
d2.nlargest(1)
```

```python
#Find out which region has more consistent with respect to GDP per capita
d3=data.groupby('Region')['GDP per capita (current US$)'].std()
print(d3)
d3.nsmallest(1)
```
