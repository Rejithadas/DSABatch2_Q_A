### 4.Read the given data set and answer the following questions. Show the steps taken to produce your final answer.

1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
2. How many patients in the data are having the blood pressure level above 120?
3. Select the rows the age is between 24 and 29 (inclusive)
4.Find the minimum and maximum value of insulin level in the data.
5. Create a histogram of "Age".
6. Create a scatter plot of "BMI" vs "Glucose".
7. Replace the min value of "BMI" with mean value of BMI
8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
import pandas as pd
df=pd.read_csv('/content/diabetes_model (1).csv')
print(df)
```

4.4
Find the minimum and maximum value of insulin level in the data.

```python
print('Insulin Minimum Value=',df['Insulin'].min())
print('Insulin Maximum Value=',df['Insulin'].max())
```

4.1Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.

```python
sample=df[45:60]
sample[['Glucose','Age']]
```

4.2 How many patients in the data are having the blood pressure level above 120?

```python
c=0
for i in df.BloodPressure:
  if i>120:
    c+=1
    print('patients having BloodPressure above 120=',c)
```

4.3
Select the rows the age is between 24 and 29 (inclusive)

```python
df[(df['Age']>=24) & (df['Age']<=29)]
```

4.4
Create a histogram of "Age".


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(df['Age'],color='green',rwidth=0.5)
plt.title(' Diabetics Patients Age Distribution',fontsize= 16)
plt.xlabel('Age')
plt.ylabel('Age Distribution')
```

4.5
 a scatter plot of "BMI" vs "Glucose".


```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
plt.title ('Scatter Plot of BMI vs Glucose')
plt.scatter(df['BMI'],df['Glucose'],s=15,c='red')
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

4.6
Replace the min value of "BMI" with mean value of BMI


```python
min_BMI = df['BMI'].min()
mean_BMI = round(df['BMI'].mean(),2)
new_df = pd.DataFrame()
new_df =df.copy() 
new_df['BMI'] = new_df['BMI'].replace(min_BMI,mean_BMI) 
new_df[new_df['BMI']==mean_BMI]
```

4.7Draw
 a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
import seaborn as sns
sns.boxplot(data=df,x='Outcome',y='Glucose')
plt.title('Distribution of Glucose ',fontsize='16')
```
