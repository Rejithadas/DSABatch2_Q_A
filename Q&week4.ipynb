```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('/content/Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
print(df)
```

```python
#Compare churn count with respect to gender
sns.set_theme(style="darkgrid")
ax = sns.countplot(y="Churn",hue="gender", data=df)
```

```python
#Find out how many female senior citizens there in the dataset
ax = sns.countplot(x="gender",hue="SeniorCitizen", data=df, palette="Set3")
```

```python
#Compare 'tenure' with 'Total Charges'
sns.lineplot(x='tenure', y='TotalCharges', data=df)
```

```python
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Contract",hue="SeniorCitizen", data=df)
```

```python
#Comment on your finds on payment method
plt.figure(figsize=(8,5))
plt.hist(df['PaymentMethod'],color='red',rwidth=1)
plt.title('Payment Methods',fontsize=16)
```
